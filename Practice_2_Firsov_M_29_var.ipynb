# Вторая часть

```python
df = pd.read_csv('/content/drive/MyDrive/PDA/Practice 2/test.csv', encoding='cp1252')
df.head()
```

```python
df = df.dropna()
df.info()
```

```python
sentiment = df['sentiment'].value_counts().head(7)
plt.bar(sentiment.index, sentiment.values, width=0.5)

plt.title('Статистика окраски твитов')
plt.xlabel('Окраска')
plt.ylabel('Количество твитов')
```

```python
data = df[df['sentiment'] == 'neutral']['Age of User'].value_counts()
plt.pie(data.values, labels=data.index, shadow=True, autopct='%.1f%%')

plt.title('Распределение твитов по возрастным группам')
```

```python
country = df['Country'].value_counts(sort=True).head(20)
plt.barh(country.index, country.values, color='b')

plt.title('Количество твитов по странам')
plt.xlabel('Количество твитов')
```

```python
fig = plt.figure()
ax = fig.add_subplot()

p11 = df[(df['Time of Tweet'] == 'morning') & (df['sentiment'] == 'positive')].shape[0]
p12 = df[(df['Time of Tweet'] == 'noon') & (df['sentiment'] == 'positive')].shape[0]
p13 = df[(df['Time of Tweet'] == 'night') & (df['sentiment'] == 'positive')].shape[0]
p21 = df[(df['Time of Tweet'] == 'morning') & (df['sentiment'] == 'neutral')].shape[0]
p22 = df[(df['Time of Tweet'] == 'noon') & (df['sentiment'] == 'neutral')].shape[0]
p23 = df[(df['Time of Tweet'] == 'night') & (df['sentiment'] == 'neutral')].shape[0]
p31 = df[(df['Time of Tweet'] == 'morning') & (df['sentiment'] == 'negative')].shape[0]
p32 = df[(df['Time of Tweet'] == 'noon') & (df['sentiment'] == 'negative')].shape[0]
p33 = df[(df['Time of Tweet'] == 'night') & (df['sentiment'] == 'negative')].shape[0]
matrix = np.array([[p11, p12, p13],
                  [p21, p22, p23],
                  [p31, p32, p33]])

b = ax.pcolormesh(matrix, edgecolors='black', cmap='plasma')
fig.colorbar(b, ax=ax)
plt.title('Распределение твитов')
plt.xlabel('Утро   -   День   -   Ночь')
plt.ylabel('Positive   -   Neutral   -   Negative')

plt.show()
```
